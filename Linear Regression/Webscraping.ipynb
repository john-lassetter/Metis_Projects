# Define webscraping functions

```python
# ex_user_agent={"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"}
```

```python
from fake_useragent import UserAgent

def get_soup(url):
    try:
        # generate random user agent for request
        user_agent = {'User-agent': UserAgent().random}
        
        response = requests.get(url, headers=user_agent)
        # 
        if response.status_code==200:
            return BeautifulSoup(response.text, 'lxml')
        else:
            print(f'Status Code Error: {response.status_code}')
    except:
        print('Failed to get response')
```

```python
def distill_soup(soup, url):
    headers=['url','address','city','state','zipcode','neighborhood','price','bdrms','baths',
                           'sq_ft','date','schools', 'transportation', 'uniques',
             'other_amenities','details','other_details']
    
    #set all features to default to nan
    address = [np.nan]*5
    rent_info = [np.nan]*4
    date = np.nan
    schools = np.nan
    transportation = np.nan
    uniques = np.nan
    other_amenities = np.nan
    details = np.nan
    other_details = np.nan
        
    #['address', 'city', 'state', 'zipcode', 'neighborhood']
    try:
        address = [x.text.replace('\n', '') for x in soup.find(class_='propertyAddressContainer').find_all('span')]
        address[-2] = address[-2][1:] #remove em dash from neighborhood
        if len(address)>2:
            del address[2]
        if len(address)>5:
            del address[-1]

        # ['price', 'bdrms', 'baths', 'sq_ft']
        rent_info = [x.text.strip('\n').strip('$') for x in soup.find(class_= 'priceBedRangeInfo').find_all(class_='rentInfoDetail')]
        feature_data.loc[0,['address','city','state','zipcode']] = rent_info
    except:
        pass
    
    # schools
    try:
        schools = []
        for table in soup.find(class_='schoolsSection', id='schoolsSection').find_all('table'):
            for row in table.find_all('tr'):
                school_dists=[]
                for column in row.find_all('td'):
                    school_dists.append(column.text.strip('\n'))
                schools.append(school_dists)

        schools = [x for x in schools if len(x)>0]
        feature_data.loc[0,['schools']] = schools
    except:
        pass
    
    # transportation
    try:
        transportation = []
        for table in soup.find(class_='profileV2TransportationSection js-viewAnalyticsSection hide-on-print',
                               id='transportationSection').find_all('table'):
            for row in table.find_all('tr'):
                trans_sub_lsts=[]
                for column in row.find_all('td'):
                    trans_sub_lsts.append(column.text.strip('\n'))
                transportation.append(trans_sub_lsts)

        transportation = [x for x in transportation if len(x)>0]
        feature_data.loc[0,['transportation']] = transportation
    except:
        pass
     
    # unique amenities
    try:
        uniques = [x.text.strip('\n') for x in soup.find_all(class_='specInfo uniqueAmenity')]
        feature_data.loc[0,['uniques']] = uniques
    except:
        pass

    # other amenities
    try:
        other_amenities = [x.text.strip('\n') for x in soup.find_all(class_='specInfo')]
        for i in uniques:
            if i in other_amenities:
                other_amenities.remove(i)
        feature_data.loc[0,['other_amenities']] = other_amenities
    except:
        pass
        
    #details including date
    try:
        details = [x.text for x in soup.find(class_='detailsContainer').find_all(class_='column')]
        
        #get date from details
        date = None
        delete_index = -1
        for i, str_ in enumerate(details):
            if 'Built in' in str_:
                date = str_.strip('Built in')
                delete_index = i
        if delete_index> -1:
            del details[delete_index] #remove date from details now that it is in its own column
            
        feature_data.loc[0,['date']] = date
        feature_data.loc[0,['details']] = details
    except:
        pass
        
    #other details (like fees and pet policies)
    try:
        other_details = []
        for container in soup.find_all(class_='feespolicies'):
            for i in range(len(container.find_all(class_='column-right'))):
                other_details.append([container.find_all(class_='column')[i].text, 
                                container.find_all(class_='column-right')[i].text])
                
        feature_data.loc[0,['other_details']] = other_details
    except:
        pass

    feature_data = ([url] + address + rent_info + [date] + schools + transportation 
             + [uniques, other_amenities, details, other_details])
   
    return pd.DataFrame.from_dict(dict(zip(headers, feature_data)), orient='index').T

```

```python
def get_data(urls, sleep_time=2):
    from time import sleep
    import random
    import datetime as dt

    df = pd.DataFrame(columns=['address','city','state','zipcode','neighborhood','price','bdrms','baths',
                               'sq_ft','date','uniques','other_amenities','details','other_details'])

    for i, url in enumerate(urls):
        soup = get_soup(url)
        df = df.append(distill_soup(soup, url), ignore_index=True)

        #save a checkpoint every 10 pages
        if i%10==9:
            checkpoint_path = 'apartments_data_checkpoint_' + (
                dt.datetime.now().strftime('%m-%d-%Y-%H-%M-%S') + '.csv')
            df.to_csv(checkpoint_path, index=False)

        # sleep to limit request rate
        sleep(sleep_time*random.random())

    return df
    
```
